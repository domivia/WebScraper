```python
url = "https://www.ceneo.pl/47293506#tab=reviews"
response = requests.get(url)
print(response.status_code)
```

```python
page_dom = BeautifulSoup(response.text, "html.parser")
print(page_dom)
```

```python
opinions = page_dom.select("div", class_="js_product-review:(.user-post--highlight)")
print(len(opinions))
print(type(opinions))
```

```python
opinion = page_dom.select_one("div", class_="js_product-review:(.user-post--highlight)")
print(type(opinion))
```

```python
opinion = opinions.pop(0)
print(type(opinion))
```
